```python
import re
import string
from collections import Counter
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def corpus(filename):
    f=open(filename, "r")
    words=[]
    lines=f.readlines()
    for line in lines:
#         words+=re.findall(r'\w', line.lower())
        line=line.lower()
        line=line.strip().split(" ")
        for i in line:
            i=''.join([j for j in i if j not in string.punctuation])
            if len(i)>0:
                words.append(i)
    return words
```

```python
corpus=corpus("dataset.txt")+corpus("austen-persuasion.txt")+corpus("austen-sense.txt")
```
